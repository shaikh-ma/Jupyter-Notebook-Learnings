```python
import pandas as pd
import numpy as np
```

```python
#a 2D array with 10 rows and 10 columns 

a_2d_array = np.random.rand(10,10)
```

## Creating the DataFrame Object from a 2d array

```python
# Converting the 2D array to Pandas DataFrame

df = pd.DataFrame(a_2d_array)
df
```

## Getting the details of the DataFrame Object

```python
# Type of the DataFrame object

type(df)
```

```python
# Data Type of a DataFrame Object

df.dtypes
```

```python
# Getting the Index

df.index
```

```python
# Changing the index

df.index = range(10,20)
```

```python
# Getting the columns

df.columns
```

```python
# Changing the columns

df.columns = list('ABCDEFGHIJ')
```

```python
# Getting first few rows

df.head(3) # Returns n rows if n is passed else prints first 5 rows
```

```python
# Getting last few rows

df.tail()
```

```python
# Getting the statistical information of the DataFrame

df.describe()
```

```python
# Getting short info

df.info()
```

```python
# Getting the counts of non-NA values for each row

df.count()
```

```python
# Getting the counts of non-NA values for each COLUMN

df.count(axis=1)
```

# Accessing Values of the DataFrame Object

```python
# Accessing entire column

print('column 1\n', df[0])
print()
print('column 2\n', df[2])
```

```python
# Accessing a single row by index name 

df.loc[10]
```

```python
# Accessing a single by index

print(df.iloc[0],end="\n")
print()
print(df.loc[12,:]) 
```

```python
# Accessing only the required cells

print(df.loc[13,["A","C"]])
```

```python
# Accessing a particular value

print(df.loc[10,'A'],     \
      df.at[10,'A']       , sep="\n")
```

# DELETING ROWS & COLUMNS

```python
# Deleting a column

df.drop('J', axis=1)
```

```python
# Deleting the first row (in this case with index 10)

df.drop(10)
```
